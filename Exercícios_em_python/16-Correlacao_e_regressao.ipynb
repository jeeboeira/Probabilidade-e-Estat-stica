```python
import matplotlib.pyplot as plt
import numpy as np

def plot_scatter(x, y, title="Gráfico de Dispersão", color="b", xlabel="X", ylabel="Y"):
    """
    Plota um gráfico de dispersão (scatter plot) para os dados fornecidos, com setas nas pontas dos eixos.
    
    Parâmetros:
    - x (array-like): valores no eixo X
    - y (array-like): valores no eixo Y
    - title (str): título do gráfico
    - color (str): cor dos pontos
    - xlabel (str): rótulo para o eixo X
    - ylabel (str): rótulo para o eixo Y
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color=color)
    plt.title(title)
    plt.grid(True)

    # Configura os limites do gráfico para deixar espaço para as setas
    x_min, x_max = min(x) + 0, max(x) + 1
    y_min, y_max = min(y) + 0, max(y) + 1
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    
    # Adiciona setas nas pontas dos eixos X e Y
    plt.annotate('', xy=(x_max, 0.1), xytext=(x_min, 0.1),
                 arrowprops=dict(arrowstyle="->", color='black', lw=0.8))
    plt.annotate('', xy=(0.1, y_max), xytext=(0.1, y_min),
                 arrowprops=dict(arrowstyle="->", color='black', lw=0.8))
    
    # Posiciona os rótulos dos eixos X e Y perto das linhas dos eixos
    plt.text(x_max, -0.5, xlabel, ha='right', va='top', fontsize=12)
    plt.text(-0.5, y_max, ylabel, ha='right', va='top', fontsize=12)
    
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_scatter(x, y, title="Gráfico de Dispersão", color="b", xlabel="X", ylabel="Y"):
    """
    Plota um gráfico de dispersão (scatter plot) para os dados fornecidos.
    
    Parâmetros:
    - x (array-like): valores no eixo X
    - y (array-like): valores no eixo Y
    - title (str): título do gráfico
    - color (str): cor dos pontos
    - xlabel (str): rótulo para o eixo X
    - ylabel (str): rótulo para o eixo Y
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color=color)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

# Exemplos de uso:
# Correlação positiva
x_pos = np.arange(10)
y_pos = x_pos + np.random.normal(0, 1, 10)
plot_scatter(x_pos, y_pos, title="Correlação Positiva", color="blue")

# Correlação negativa
x_neg = np.arange(10)
y_neg = -x_neg + np.random.normal(0, 1, 10)
plot_scatter(x_neg, y_neg, title="Correlação Negativa", color="red")

# Nenhuma correlação
x_none = np.arange(10)
y_none = np.random.normal(0, 5, 10)
plot_scatter(x_none, y_none, title="Nenhuma Correlação", color="green")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_scatter_with_origin_detail(x, y, title="Gráfico de Dispersão", color="b", xlabel="X", ylabel="Y"):
    """
    Plota um gráfico de dispersão e adiciona um detalhe visual para indicar que o gráfico não começa no ponto 0.
    
    Parâmetros:
    - x (array-like): valores no eixo X
    - y (array-like): valores no eixo Y
    - title (str): título do gráfico
    - color (str): cor dos pontos
    - xlabel (str): rótulo para o eixo X
    - ylabel (str): rótulo para o eixo Y
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color=color)
    plt.title(title)
    plt.grid(True)

    # Configura os limites do gráfico para deixar espaço para as setas
    x_min, x_max = min(x) - 1, max(x) + 1
    y_min, y_max = min(y) - 1, max(y) + 1
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)

    # Adiciona setas nas pontas dos eixos X e Y
    plt.annotate('', xy=(x_max, 0), xytext=(x_min, 0),
                 arrowprops=dict(arrowstyle="->", color='black', lw=1))
    plt.annotate('', xy=(0, y_max), xytext=(0, y_min),
                 arrowprops=dict(arrowstyle="->", color='black', lw=1))
    
    # Adiciona um marcador ou linha indicando que o gráfico não começa no ponto (0, 0)
    plt.axhline(0, color='black', linestyle='--', lw=1)  # Linha horizontal tracejada no 0 no eixo Y
    plt.axvline(0, color='black', linestyle='--', lw=1)  # Linha vertical tracejada no 0 no eixo X
    
    # Marcador no ponto onde o gráfico começa (não na origem)
    plt.text(x_min, 0, "Ponto de início", ha='right', va='top', fontsize=10, color="red")
    
    # Posiciona os rótulos dos eixos X e Y perto das linhas dos eixos
    plt.text(x_max, -0.5, xlabel, ha='right', va='top', fontsize=12)
    plt.text(-0.5, y_max, ylabel, ha='right', va='top', fontsize=12)
    
    plt.show()

# Exemplo de uso:
# Dados do gráfico (modifique conforme necessário)
x = np.array([16, 25, 39, 45, 49, 64, 70, 29, 57, 22])
y = np.array([109, 122, 143, 132, 199, 185, 199, 130, 175, 118])

plot_scatter_with_origin_detail(x, y, title="Gráfico com Detalhe de Ponto de Início", color="blue", xlabel="Eixo X", ylabel="Eixo Y")
```
